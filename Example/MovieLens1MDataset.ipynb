```python
pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('/Users/admin/Workspace/data-analysis/DataLoading/datasets/movielens/users.dat', sep='::',
                      header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('/Users/admin/Workspace/data-analysis/DataLoading/datasets/movielens/ratings.dat', sep='::',
                        header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('/Users/admin/Workspace/data-analysis/DataLoading/datasets/movielens/movies.dat', sep='::',
                       header=None, names=mnames)
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
data.iloc[0]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings
```
