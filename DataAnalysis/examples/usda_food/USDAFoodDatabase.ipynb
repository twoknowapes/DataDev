```python
nutrients = []

for rec in db:
    fnuts = pd.DataFrame(rec['nutrients'])
    fnuts['id'] = rec['id']
    nutrients.append(fnuts)

nutrients = pd.concat(nutrients, ignore_index=True)
nutrients
```

```python
nutrients.duplicated().sum()  # 重复数据数量
nutrients = nutrients.drop_duplicates()  # 删除重复数据
```
